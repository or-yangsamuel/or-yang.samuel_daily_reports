### Exercise 9:
Create a subplot with 2 rows and 1 column. Plot `y1 = x^2` in the first subplot and `y2 = x^3` in the second.

```python
# Exercise 9

````python
x=np.linspace(0,10)

fig,axes=plt.subplots(1,2,figsize=(8,6),dpi=100)
y1=x**2
y2=x**3
#ploting subplot 1
axes[0].plot(x,y1,c='r',label=r'y1=x^2')
axes[0].plot(x,y2,c='g',label=r'y2=x^3')
axes[0].legend(loc=2)
axes[0].set_xlabel('x')
axes[0].set_ylabel('y1,y2')
axes[0].set_title(r'y_1=x^2 and y_2=x^3')
#ploting subplot2

axes[1].plot(x,y1,c='r',label=r'y1=x^2')
axes[1].plot(x,y2,c='g',label=r'y2=x^3')
axes[1].legend(loc=2)
axes[1].set_xlabel('x')
axes[1].set_ylabel('y1,y2')
axes[1].set_title(r'y_1=x^2 and y_2=x^3')
fig.tight_layout()
plt.show()
````

Plot a filled area under the curve of `y = x^2` for `x` ranging from 0 to 10.

```python
# Exercise 14

````python
x=np.linspace(0,10)
y=x**2
plt.plot(x,y,label=r'y=x^2')
plt.fill_between(x,y,color='g',alpha=0.4)
plt.legend()
#plt.xlabel('x')
plt.ylabel('y')
plt.show()
````
